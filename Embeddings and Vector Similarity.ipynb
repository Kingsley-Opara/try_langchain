```python
import google.generativeai as genai

genai.configure(api_key = GEMINI_API_KEY)
```

```python
def get_embedding(text, model = "models/text-embedding-004"):
    result = genai.embed_content(
        model=model,
        content=text,
        task_type="retrieval_document",
        title="Embedding of single string")
    return result
# 'all-mpnet-base-v2'
# "all-MiniLM-L6-v2"

# def get_embedding(text, model = SentenceTransformer("all-mpnet-base-v2")):
#     result = model.encode(text)
#     return result
```

```python
documents = [
    "The cat jmped over the dog",
    "The red cat jmped over the dog",
    "The turkey ran in circles",
   
]
```

```python
embeddings = [get_embedding(x) for x in documents]
```

```python
def calculate_cosine_metrics(v1, v2):
    dot_product = np.dot(v1, v2)
    magnitude1 = np.linalg.norm(v1)
    magnitude2 = np.linalg.norm(v2)
    cosine_simlarity = dot_product/(magnitude1 * magnitude2)
    cosine_distance = 1 - cosine_simlarity
    return int(cosine_simlarity * 100), int(cosine_distance * 100)

query_embedding = get_embedding("Obi cleaned the house and did his homework")

query_embedding['embedding']
    
```

```python
# print(calculate_cosine_metrics(embeddings[0]['embedding'], embeddings[0]['embedding'])
print(calculate_cosine_metrics(embeddings[0]['embedding'], query_embedding['embedding']))
print(calculate_cosine_metrics(embeddings[0]['embedding'], embeddings[1]['embedding']))
print(calculate_cosine_metrics(embeddings[1]['embedding'], embeddings[2]['embedding']))
```
